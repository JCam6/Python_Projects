```python
import pandas as pd
from sklearn.model_selection import train_test_split
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
# Read in training csv into dataframe
df = pd.read_csv('train.csv')

# Split dataset
train_data, test_data = train_test_split(df, test_size=0.1, shuffle=True, random_state=1)

# Create variable incase only modeling on part of dataset 
k = train_data.shape[0] # grab the whole dataset

# Fit AutoML to data set and identify classification feature (e.g. Cliked)
predictor = TabularPredictor(label='Clicked').fit(train_data.head(k))

#predictor.fit_summary()

# Create Predictions
y_pred = predictor.predict(test_data.head(k))

# Evaluate
predictor.evaluate_predictions(y_true=test_data['Clicked'], y_pred = y_pred, auxiliary_metrics = True)
```

```python
# View performance of models
predictor.leaderboard(test_data, silent=True)
```
